# Welcome to the Anime Rating Comparator!

This project is designed to compare various anime ratings to other data points, such as number of user votes, format category, and more. 

```python
# Library import statements
import sqlite3
import pandas as pd
from matplotlib import pyplot as plt
```

```python
# Uses pandas to read dataset and stores via python as a variable
anime_data = pd.read_csv("anime.csv")
```

Here is the main data set used for this project. This data comes from the myanimelist.net API, a collection of data pulled from 12294 anime titles and 73516 users of one of the leading anime-related database sites on the internet. Though slightly outdated, the data from this API still overall accurately reflects the current state of the various title ratings, especially when user votes are into the thousands.

You can view the original data set here: https://www.kaggle.com/CooperUnion/anime-recommendations-database

For purposes of having the most rounded possible view of the data, titles with less than 1000 user votes are still included in the queries, graphs, and analyses, as there are enough of these titles in the main data set to not be considered outliers in individual categorical analysis. However, anything tagged as adult-oriented has been excluded and removed from the original data set, as these points do act as outliers, and are not necessary for a proper analysis of the data. This exclusion reduces the total number of data points in the population to 11153, which is still perfectly viable for these analyses.

```python
# Main data table
anime_data
```

```python
# Allows the data to be accessed via SQL queries
anime_data.to_sql("anime_data_sql", sqlite3.connect("anime.db"), if_exists = "replace")
```

**Correlations between Type and User Votes**

We will begin with User Votes. Is there a correlation between higher ratings and the amount of users that vote on something?

Does this correlation change between the four main anime formats: TV Series, Movie, OVA, and Special?

Let's check out TV Series first. For this data set, I have ommitted any entry with an unknown number of episodes. Here are the top five data points as ordered by Rating.

```python
anime_tv = sqlite3.connect("anime.db")
anime_data = pd.read_sql_query("SELECT name, type, rating, members AS 'user votes' FROM anime_data_sql WHERE type = 'TV' AND episodes != 'Unknown' ORDER BY rating DESC", anime_tv)
anime_data.head()
```

And here are the top five data points as ordered by User Votes. 

```python
anime_tv = sqlite3.connect("anime.db")
anime_data = pd.read_sql_query("SELECT name, type, rating, members AS 'user votes' FROM anime_data_sql WHERE type = 'TV' AND episodes != 'Unknown' ORDER BY members DESC", anime_tv)
anime_data.head()
```

This scatter plot displays the Ratings of TV Series against the number of users who voted for them. The y-axis is on a logarithmic scale as to better accentuate the correlation. This y-axis format applies to all following scatter plots as well.

```python
plt.figure(figsize=(15, 10))

rating = anime_data['rating']
user_votes = anime_data['user votes']

plt.scatter(rating, user_votes)

plt.yscale('log')

plt.title('Ratings vs. Votes: TV Series', fontsize=18)
plt.xlabel('Rating', fontsize=14)
plt.ylabel('User Votes', fontsize=14)
```

There is a strong positive correlation between TV Series Ratings and their User Votes.

**Results: Type vs. User Votes**

For all four type categories, there is a strong positive correlation between the average title Rating and the number of users who voted on them. What I found especially interesting, though, is that the outlier patterns are all the same - in every case, there are some titles that have a small minority of highly dedicated fans, though there are not nearly as many User Votes on them as the more popular and mainstream titles. This affirms that the quality of an anime title is not necessarily determined by the size of its fanbase, especially when the fanbase is highly dedicated to the title.
